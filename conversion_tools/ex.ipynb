```python
# for meta data

def parse_json(data_path):
    with open(data_path, 'rb') as g:
        for l in g:
            yield eval(l)

i = 0
df = {}
for d in parse_json('./meta_Musical_Instruments.json'):
    df[i] = d
    i += 1
b = pd.DataFrame.from_dict(df, orient='index')
```

```python
new_li = set()

for li in b['categories'].apply(lambda x: [e for i in x for e in i]).to_list() :
    for i in li :
        new_li.add(i)
```
